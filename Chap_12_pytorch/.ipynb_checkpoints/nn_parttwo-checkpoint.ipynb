```python
import numpy as np

import matplotlib.pyplot as plt
x_train = np.arange(10, dtype=np.float32).reshape((10, 1))
y_train = np.array([1.0, 1.3, 2.0, 2.3, 2.9, 5.0, 6.3, 6.6, 6.8, 7.4], dtype = np.float32)
plt.figure(1)
plt.plot(x_train, y_train, 'o', markersize = 10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
